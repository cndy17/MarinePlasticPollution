```python
import pandas as pd
import numpy as np
import os
```

```python
wd = '/Users/cindy/Desktop/Plastic Data/plastic-leakage-to-aquatic-environments/'
os.chdir(wd)
```

```python
df = pd.read_csv('plastic-leakage-to-aquatic-environments.csv')
```

```python
df.head()
```

```python
df_melted = pd.melt(
    df,
    id_vars=['Entity', 'Code', 'Year'],
    value_vars=[
        'Plastic leakage to aquatic envrionment - Leakage type: Accumulated stock in oceans',
        'Plastic leakage to aquatic envrionment - Leakage type: Accumulated stock in rivers and lakes'
    ],
    var_name='Body of Water',
    value_name='Plastic Leakage'
)

# Clean up the 'Body of Water' column
df_melted['Body of Water'] = df_melted['Body of Water'].str.extract(r'Accumulated stock in (.+)$')

df_melted.head()
```

```python
df_melted['Body of Water'].value_counts()
```

```python
df_melted.to_csv('plastic_leakage_data.csv', index=False)
```

Pieces/10 mins in the context of microplastics refers to a unit of measurement used to quantify the concentration of microplastic debris in a sample, specifically when collected from beaches using a 10-minute "Nurdle Patrol" method. This method involves one person hand-collecting plastic pieces from a beach area over a 10-minute period. The count is then used to compare different beach locations, providing a standardized measure of microplastic abundance. 

"Pieces kg⁻¹ d.w." is a unit of measurement for microplastic concentration, meaning the number of microplastic particles per kilogram of dry weight of a sample.
